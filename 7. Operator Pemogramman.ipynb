### 4.OPERATOR LOGIKA
Digunakan untuk membandingkan dua ekspresi/ungkapan boolean dan hasilnya adalah TRUE atau FALSE





#### Contoh Program 

```python
#And

# Pernyataan 1: 
Mahasiswa_istek_programmer = True

# Pernyataan 2: 
Mahasiswa_istek_rajin_mandi = True

Mahasiswa_istek_programmer and Mahasiswa_istek_rajin_mandi
```

```python
## OR

# Pernyataan 1: 
Mahasiswa_istek_programmer = True

# Pernyataan 2: 
Mahasiswa_istek_rajin_mandi = False

Mahasiswa_istek_programmer or Mahasiswa_istek_rajin_mandi
```

```python
## OR

# Pernyataan 1: 
Mahasiswa_istek_programmer = True

# Pernyataan 2: 
Mahasiswa_istek_rajin_mandi = False

not Mahasiswa_istek_rajin_mandi
```

```python
## Not

# Pernyataan 1: 
Mahasiswa_istek_programmer = True

# Pernyataan 2: 
Mahasiswa_istek_rajin_mandi = False

not Mahasiswa_istek_rajin_mandi
```
